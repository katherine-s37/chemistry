```python
import pandas as pd
from mp_api.client import MPRester
from pymatgen.analysis.local_env import CrystalNN
from pymatgen.core import Structure
```

```python
# 示例数据：催化剂名称
data = {
    'catalyst': [
        'Fe3O4',  # 氧化铁
        'ZnO'  # 氧化锌
    ]
}
df = pd.DataFrame(data)
```

```python
# 假设你已经有了这些唯一值
family_unique = ['In2O3', 'Cu', 'Pd', 'ZnO-ZrO2']
support_unique = ['Fe3O4', 'ZnO', 'SBA-15', 'MCM-15', 'CeO2', '0', 'ZrO2', 'TiO2', 'AlCeO', 'SiO2', 'CNT', 'Al2O3', 'In2O3', 'ZnTiO3', 'Ga2O3', 'SiC']
promoter_unique = ['Ni', 'Mn', '0', 'Cu', 'Pt', 'Ir', 'Ga', 'Y', 'Ni', 'Ce', 'Pd', 'Ru', 'Rh', 'In', 'Zr', 'Pr', 'Ag', 'Ce', 'La', 'Graphene (C)', 'Mn', 'Nd', 'Zn', 'La', 'Pd', 'Al', 'Si', 'N2', 'Mg', 'W', 'Ba', 'Mo', 'Ga', 'Cr', 'N doped Graphene(N)', 'K', 'Cu', 'Zn', 'Mg', 'Fe']

# 用于存储描述符的空字典
descriptors = {
    'Family': [],
    'Support': [],
    'Promoter': [],
    'Descriptor': []
}
```

```python
def get_material_descriptors(material_name):
    try:
        # 查询材料属性
        material_data = mpr.get_data(material_name)
        if material_data:
            # 提取感兴趣的描述符
            material_info = material_data[0]
            # 选择一些常见的描述符
            descriptors = {
                'band_gap': material_info.get('band_gap', None),  # 能带隙
                'density': material_info.get('density', None),  # 密度
                'volume': material_info.get('volume', None),  # 体积
                'formation_energy_per_atom': material_info.get('formation_energy_per_atom', None),  # 每个原子的形成能
                'spacegroup': material_info.get('spacegroup', {}).get('symbol', None),  # 空间群符号
                'total_magnetization': material_info.get('total_magnetization', None),  # 总磁化强度
                'elasticity': material_info.get('elasticity', {}).get('bulk_modulus', None),  # 体积模量
            }
            return descriptors
        else:
            return None
    except:
        return None
```

```python
# 获取 Family 列的描述符
for material in family_unique:
    descriptor = get_material_descriptors(material)
    descriptors['Family'].append(material)
    descriptors['Support'].append(None)
    descriptors['Promoter'].append(None)
    descriptors['Descriptor'].append(descriptor)

# 获取 Support 列的描述符
for material in support_unique:
    descriptor = get_material_descriptors(material)
    descriptors['Family'].append(None)
    descriptors['Support'].append(material)
    descriptors['Promoter'].append(None)
    descriptors['Descriptor'].append(descriptor)

# 获取 Promoter 列的描述符
for material in promoter_unique:
    descriptor = get_material_descriptors(material)
    descriptors['Family'].append(None)
    descriptors['Support'].append(None)
    descriptors['Promoter'].append(material)
    descriptors['Descriptor'].append(descriptor)

# 将描述符存入数据框中
descriptor_df = pd.DataFrame(descriptors)

# 输出结果数据框
print(descriptor_df)

# 将结果存入 CSV 文件
descriptor_df.to_csv("material_descriptors.csv", index=False)
```

```python
def get_material_properties(material):
    try:
        print(f"Fetching data for: {material}")  # 调试输出
        # 使用化学系统进行搜索
        results = mpr.summary.search(chemsys=material)
        print(f"Results: {results}")  # 调试输出
        if results:
            # 获取第一个匹配的材料
            material_data = results[0]
            print(f"Material Data: {material_data}")  # 调试输出
            # 获取3D结构
            structure_dict = material_data.structure.as_dict()
            print(f"Structure Dict: {structure_dict}")  # 调试输出
            structure = Structure.from_dict(structure_dict)
            # 计算局部环境（例如，使用CrystalNN）
            nn = CrystalNN()
            local_env = nn.get_all_nn_info(structure)
            # 返回3D描述符（示例：局部环境和体积）
            return {
                'volume': structure.volume,
                'local_env': local_env
            }
        else:
            return None
    except Exception as e:
        print(f"Error fetching data for {material}: {e}")
        return None
```

```python
# 转换材料名称为化学系统字符串（例如，将Fe3O4转换为'Fe-O'）
def convert_to_chemsys(material_name):
    elements = []
    for char in material_name:
        if char.isupper():
            elements.append(char)
        elif char.islower():
            elements[-1] += char
        elif char.isdigit():
            continue  # 跳过数字部分
    return '-'.join(sorted(set(elements)))

# 转换材料名称为化学系统
df['chemsys'] = df['catalyst'].apply(convert_to_chemsys)
```

```python
# 转换材料名称为化学系统
df['chemsys'] = df['catalyst'].apply(convert_to_chemsys)

# 获取无机物的3D描述符
properties_list = df['chemsys'].apply(get_material_properties)
properties_df = pd.DataFrame(properties_list.tolist())

# 将描述符和原始数据合并
result_df = pd.concat([df, properties_df], axis=1)

print(result_df)
```
